```python
### 2. Prepare data ###

import os
import random
from PIL import Image, ImageDraw, ImageFont, ImageFilter
import numpy as np

# ========== CONFIG ==========
SAVE_DIR = "/content/data"
FONT_DIR = "/content/fonts"  # upload fonts here
os.makedirs(SAVE_DIR, exist_ok=True)

# words for generating random sentences
WORDS = [
    "hello", "world", "apple", "robot", "code", "deep", "learning", "machine",
    "text", "ocr", "easy", "pytorch", "tesseract", "google", "amazon", "detect"
]

# load ttf fonts
fonts = [os.path.join(FONT_DIR, f) for f in os.listdir(FONT_DIR) if f.endswith(".ttf")]
assert len(fonts) > 0, "‚ùó ËØ∑ÂÖà‰∏ä‰º†Â≠ó‰ΩìÂà∞ /content/fonts Êñá‰ª∂Â§πÔºÅ"


def random_sentence(min_words=2, max_words=4):
    """randomly choose English words to form a sentence"""
    k = random.randint(min_words, max_words)
    return " ".join(random.choices(WORDS, k=k))


def generate_image(text, save_path):
    """create text image with noise & font randomness"""
    W, H = 400, 120
    img = Image.new("RGB", (W, H), color=(255, 255, 255))
    draw = ImageDraw.Draw(img)

    # random font and size
    font_path = random.choice(fonts)
    font = ImageFont.truetype(font_path, random.randint(25, 38))

    # random text position
    x = random.randint(10, 40)
    y = random.randint(20, 50)
    draw.text((x, y), text, fill=(0, 0, 0), font=font)

    # random blur/noise
    if random.random() < 0.5:
        img = img.filter(ImageFilter.GaussianBlur(radius=random.uniform(0.3, 1.0)))

    img.save(save_path)


# ====== generate 100 images ======
for i in range(100):
    sentence = random_sentence()
    filename = sentence.replace(" ", "_") + ".png"
    save_path = os.path.join(SAVE_DIR, filename)
    generate_image(sentence, save_path)

print("üéâ Synthetic dataset created at /content/data (100 images)")
```
